```python
import pandas as pd
```

## Scraping 

```python
df=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")[0]
```

```python
df
```

### Dropping Not Assigned in Borough

```python
df=df[df.Borough!='Not assigned']
df
```

### Grouping duplicate Postal Code

```python
df=df.groupby(['Postal Code','Borough'])['Neighbourhood'].apply(','.join).reset_index()
df
```

### Changing 'Not assigned' Neighbourhood to it's Borough

```python
import numpy as np
df['Neighbourhood'] = np.where(df['Neighbourhood'] == 'Not assigned', df['Borough'], df['Neighbourhood'])
```

```python
df.shape
```

## Adding latitude and Longitude

```python
dfll=pd.read_csv("http://cocl.us/Geospatial_data")
```

```python
dfll.head()
```

```python
df = pd.merge(df, dfll, on='Postal Code')
df
```

### Finding Toronto Latitude and Longitude using Geopy

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(df['Latitude'], df['Longitude'], df['Neighbourhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

## Clustering by Distance using KKN

```python
k_means = KMeans(init="k-means++", n_clusters=4, n_init=12)
```

### making vectors from dataframe

```python
dfn=df.drop(columns=["Postal Code","Borough","Neighbourhood"])
dfn
```

```python
X=dfn.values
```

```python
k_means.fit(X)
```

```python
k_means_labels = k_means.labels_
k_means_labels
```

```python
k_means_cluster_centers = k_means.cluster_centers_
k_means_cluster_centers
```

### Making a Map

```python
#adding label to dataframe
df["Label"]=k_means_labels
df
```

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label, delta in zip(df['Latitude'], df['Longitude'], df['Neighbourhood'], df['Label']):
    if delta==1:
        label = folium.Popup(label, parse_html=True)
        folium.CircleMarker(
            [lat, lng],
            radius=5,
            popup=label,
            color='blue',
            fill=True,
            fill_color='#3186cc',
            fill_opacity=0.7,
            parse_html=False).add_to(map_toronto)
    elif delta==2:
        label = folium.Popup(label, parse_html=True)
        folium.CircleMarker(
            [lat, lng],
            radius=5,
            popup=label,
            color='yellow',
            fill=True,
            fill_color='#FF0000',
            fill_opacity=0.7,
            parse_html=False).add_to(map_toronto)
    elif delta==3:
        label = folium.Popup(label, parse_html=True)
        folium.CircleMarker(
            [lat, lng],
            radius=5,
            popup=label,
            color='red',
            fill=True,
            fill_color='#FF0000',
            fill_opacity=0.7,
            parse_html=False).add_to(map_toronto)
    else:
        label = folium.Popup(label, parse_html=True)
        folium.CircleMarker(
            [lat, lng],
            radius=5,
            popup=label,
            color='green',
            fill=True,
            fill_color='#00FF00',
            fill_opacity=0.7,
            parse_html=False).add_to(map_toronto) 
map_toronto
```
